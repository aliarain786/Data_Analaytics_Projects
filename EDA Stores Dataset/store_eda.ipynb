## **Begin With The visualization Now**

```python
plt.figure(figsize=(19,10))
sns.heatmap(df.corr(), annot=True)
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(16,8))
fig1 =sns.barplot(x='Region', y='Sales', data=df)
fig1.set_title('Sales by Region', fontsize=30, fontweight='bold', color='blue')
fig1.set_xlabel('Region', fontsize=24, fontweight='bold', color='blue')
fig1.set_ylabel('Sales', fontsize=24, fontweight='bold', color='blue')
fig1.tick_params(labelsize=12)
```

```python
plt.figure(figsize=(14,8))
fig2=sns.lineplot(x='Discount', y='Sales', data=df)
fig2.set_title('Sales by Discount', fontsize=35, fontweight='bold', color='blue')
fig2.set_xlabel('Discount', fontsize=24, fontweight='bold', color='blue')
fig2.set_ylabel('Sales', fontsize=24, fontweight='bold', color='blue')
fig2.tick_params(labelsize=24)
```

```python
plt.figure(figsize=(14,8))
fig3= sns.barplot(x='Ship Mode', y='Sales', data=df)
fig3.set_title('Sales by Ship Mode', fontsize=35, fontweight='bold', color='blue')
fig3.set_xlabel('Ship Mode', fontsize=24, fontweight='bold', color='blue')
fig3.set_ylabel('Sales', fontsize=24, fontweight='bold', color='blue')
fig3.tick_params(labelsize=24)
```

```python
plt.figure(figsize=(14,8))
fig4 =sns.histplot(x='Market', data= df, bins=7, kde=True, color='blue')
fig4.set_title('Market', fontsize=45, fontweight='bold', color='blue')
fig4.set_xlabel('Market', fontsize=35, fontweight='bold', color='blue')
fig4.set_ylabel('Sales', fontsize=35, fontweight='bold', color='blue')
fig4.tick_params(labelsize=24)
```

```python
group1 = df.groupby('Category')['Sales'].sum().reset_index()
group1
```

```python
fig5 = px.pie(group1, values='Sales', names='Category', title='Sales by Category')
fig5.update_traces(textposition='inside', textinfo='percent+label')
fig5.update_layout(title_x=0.5, title_font_size=30)
fig5.update_traces(textfont_size=30)
```

```python
group2 = df.groupby('Country')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
group2
```

```python
fig6 = px.bar(group2.head(5) , x='Country', y='Sales', color='Country' , title='Sales by Country')
fig6.update_layout(title_x=0.5, title_font_size=30)
fig6.update_xaxes(title_text='Country', title_font_size=24, title_font_color='blue')
fig6.update_yaxes(title_text='Sales', title_font_size=24, title_font_color='blue')
fig6.update_xaxes(tickfont_size=20)
```

```python
fig7 =px.scatter(df, x='Order Date', y='Profit',color='Discount' ,title='Sales over time')
fig7.update_layout(title_x=0.5, title_font_size=30)
fig7.update_xaxes(title_text='Order Date', title_font_size=24, title_font_color='blue')
fig7.update_yaxes(title_text='Profit', title_font_size=24, title_font_color='blue')
fig7.update_xaxes(tickfont_size=20)
fig7.update_yaxes(tickfont_size=20)
```
